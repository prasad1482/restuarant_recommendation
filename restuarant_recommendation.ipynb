```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Dataset .csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
most_common_cuisine = df['Cuisines'].mode()[0]
df['Cuisines'].fillna(most_common_cuisine, inplace=True)
```

```python
df.isnull().sum()
```

```python
# 📊 Unique Cuisines
print("\nUnique Cuisine Types:")
print(df['Cuisines'].unique())
```

```python
# 📈 Rating Statistics
print("\nAggregate Rating Stats:")
print(df['Aggregate rating'].describe())
```

#content based filtering


```python
favorite_cuisine = 'Italian'
price_range = 3
city = 'Delhi'
```

```python
# Filter Based on Cuisine
recommended_restaurants = df[df['Cuisines'].str.contains(favorite_cuisine, case=False, na=False)]
```

```python
# Further Filter by Price Range
recommended_restaurants = recommended_restaurants[recommended_restaurants['Price range'] == price_range]
```

```python
# Further Filter by City
recommended_restaurants = recommended_restaurants[recommended_restaurants['City'].str.contains(city, case=False, na=False)]
```

```python
# Sort by Ratings
recommended_restaurants = recommended_restaurants.sort_values(by='Aggregate rating', ascending=False)
```

```python
print("\n📌 Content-Based Recommendations:")
print(recommended_restaurants[['Restaurant Name', 'Cuisines', 'City', 'Price range', 'Aggregate rating']])
```

#Collaborative Filtering

```python
# Create Pivot Table
user_restaurant_matrix = df.pivot_table(index='City', columns='Restaurant Name', values='Aggregate rating')
```

```python
# Fill NA with 0 (or you can try mean instead)
user_restaurant_matrix.fillna(0, inplace=True)
```
