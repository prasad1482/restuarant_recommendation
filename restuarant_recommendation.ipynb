```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Dataset .csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
most_common_cuisine = df['Cuisines'].mode()[0]
df['Cuisines'].fillna(most_common_cuisine, inplace=True)
```

```python
df.isnull().sum()
```

#content based filtering


```python
unique_cusines=df["Cuisines"].unique()
print(unique_cusines)
```

```python
  favorite_cuisine = 'Italian'
  recommended_restaurants = df[df['Cuisines'].str.contains(favorite_cuisine, case=False, na=False)]
  print(recommended_restaurants[['Restaurant Name', 'Cuisines', 'City']])
```

```python
user_price_range = 3  # Example: 1 (low) to 4 (high)
recommended_restaurants = df[
    (df['Cuisines'].str.contains(favorite_cuisine, case=False, na=False)) &
    (df['Price range'] == user_price_range)
]
print(recommended_restaurants[['Restaurant Name', 'Cuisines', 'City', 'Price range', 'Aggregate rating']])
```

```python
#sorted recommendation
recommended_restaurants = recommended_restaurants.sort_values(by='Aggregate rating', ascending=False)
print(recommended_restaurants[['Restaurant Name', 'Cuisines', 'Aggregate rating', 'Votes']])
```
